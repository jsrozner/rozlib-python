```python

from conllu import TokenList
from lib.plotting.graph_affinity import draw_custom_dependency_graph


def get_head_dep_list(tl: TokenList):
    print(tl)
    sent_words = list(map(lambda x: x['form'], tl))
    edges = []
    node_vals = []
    # print(tl.metadata['text'])
    for i, t in enumerate(tl):
        print(t['head'])
        node_vals.append(1)
        head_idx = t['head'] - 1
        if head_idx >= 0:
            edges.append((i, head_idx, 1))
        # head = tl[head_idx]
        # print(f"{i}:{t['form']}:{head}")
    
    draw_custom_dependency_graph(
        sent_words,
        edges,
        node_vals,
    )

get_head_dep_list(all_data_parsed[16])
```
